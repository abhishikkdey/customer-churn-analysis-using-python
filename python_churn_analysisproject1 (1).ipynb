```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# List of service-related columns
services = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV'
]

# Melt the dataframe to long format
df_melted = df.melt(id_vars='Churn', value_vars=services,
                    var_name='Service', value_name='Status')

# Group by service, status, and churn
grouped = df_melted.groupby(['Service', 'Status', 'Churn']).size().reset_index(name='Count')

# Plot
plt.figure(figsize=(16, 8))
sns.barplot(
    data=grouped,
    x='Service',
    y='Count',
    hue='Churn',
    palette='Set2'
)

plt.title('Customer Churn by Service Subscription')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Use your existing churn count table
churn_counts = df.groupby('PaymentMethod')['Churn'].value_counts().unstack().fillna(0)

# Plot
churn_counts.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# Labels and title
plt.title('Churn by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=90)
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```
